```
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

Calculating sentiment score for positive polarity files :

```
poslist = []

def sentiment_scores(text): 
   
    sid_obj = SentimentIntensityAnalyzer() 
  
    sentiment_dict = sid_obj.polarity_scores(text) 
      
    print("Overall sentiment dictionary is : ", sentiment_dict) 
    print("file was rated as ", sentiment_dict['neg']*100, "% Negative") 
    print("file was rated as ", sentiment_dict['neu']*100, "% Neutral") 
    print("file was rated as ", sentiment_dict['pos']*100, "% Positive") 
  
    print("File Overall Rated As", end = " ") 

    if sentiment_dict['compound'] >= 0.05 : 
      print("Positive") 
      poslist.append(sentiment_dict['pos'])

    elif sentiment_dict['compound'] > -0.05 and sentiment_dict['compound'] < 0.05 : 
      print("Neutral") 

    elif sentiment_dict['compound'] <= - 0.05 : 
      print("Negative") 

    print("\n")  
```

```
for f in txt :
  print("File : ",f)
  m = open(os.path.join(path,f), 'r').read()
  sentiment_scores(m)
```

```
def Average(list1): 
    return sum(list1) / len(list1)
```

```
average = Average(poslist)
print("The threshold required for positive files are : ",average)
```

Calculating Sentiment score for negative polarity files :

```
neglist = []

def sentiment_scores1(text): 
   
    sid_obj1 = SentimentIntensityAnalyzer() 
  
    sentiment_dict1 = sid_obj1.polarity_scores(text) 
      
    print("Overall sentiment dictionary is : ", sentiment_dict1) 
    print("file was rated as ", sentiment_dict1['neg']*100, "% Negative") 
    print("file was rated as ", sentiment_dict1['neu']*100, "% Neutral") 
    print("file was rated as ", sentiment_dict1['pos']*100, "% Positive") 
  
    print("File Overall Rated As", end = " ") 

    if sentiment_dict1['compound'] >= 0.05 : 
      print("Positive")     

    elif sentiment_dict1['compound'] > -0.05 and sentiment_dict1['compound'] < 0.05 : 
      print("Neutral") 

    elif sentiment_dict1['compound'] <= - 0.05 : 
      print("Negative") 
      neglist.append(sentiment_dict1['neg'])

    print("\n")  
```
